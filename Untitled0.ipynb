Actividad   
1: Juego de adivinar el número Pasos:
1. Importar el módulo random para generar números aleatorios.
2. Crear una función generar_numero() que retorne un número aleatorio entre 1 y 100.
3. Implementar un bucle while para el juego principal.
4. Dentro del bucle, usar try-except para manejar entradas no numéricas.
5. Usar if-elif-else para dar pistas al jugador (mayor, menor o correcto).
6. Implementar un contador de intentos.
7. Usar break para salir del bucle cuando se adivine el número.

```python
import random

def generar_numero():
    # Paso 2: Retorna un número aleatorio entre 1 y 100
    return random.randint(1, 100)

def jugar():
    numero_secreto = generar_numero()
    intentos = 0
    print("¡Bienvenido al juego de adivinar el número!")
    print("He pensado un número entre 1 y 100. ¿Puedes adivinarlo?")

    # Paso 3: Bucle principal del juego
    while True:
        # Paso 4: Manejo de entradas no numéricas
        try:
            intento_usuario = int(input("\nIntroduce tu número: "))
            intentos += 1 # Paso 6: Contador de intentos
        except ValueError:
            print("⚠️ Error: Por favor, introduce un número válido (entero).")
            continue

        # Paso 5: Pistas al jugador
        if intento_usuario < numero_secreto:
            print("El número es mayor ↑")
        elif intento_usuario > numero_secreto:
            print("El número es menor ↓")
        else:
            # Paso 7: Salir del bucle al adivinar
            print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
            break

# Ejecutar el juego
if __name__ == "__main__":
    jugar()
```

Actividad 2: Calculadora de operaciones básicas Pasos:
1. Definir funciones para cada operación (suma, resta, multiplicación, división).
2. Crear un menú de opciones usando un bucle while.
3. Utilizar if-elif-else para seleccionar la operación.
4. Implementar try-except para manejar errores (como división por cero).
5. Usar continue para volver al menú principal después de cada operación.
6. Implementar una opción para salir del programa usando break.

```python
# 1. Definición de funciones para cada operación
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    # 4. Manejo de error específico (División por cero)
    if b == 0:
        raise ZeroDivisionError("No se puede dividir entre cero.")
    return a / b

# 2. Bucle principal para el menú
while True:
    print("\n--- CALCULADORA BÁSICA ---")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Salir")

    opcion = input("\nSelecciona una opción (1-5): ")

    # 6. Opción para salir del programa
    if opcion == '5':
        print("¡Hasta luego, Luis Alfonso!")
        break

    # 3. Selección de la operación con if-elif-else
    if opcion in ['1', '2', '3', '4']:
        try:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))

            if opcion == '1':
                print(f"Resultado: {num1} + {num2} = {sumar(num1, num2)}")
            elif opcion == '2':
                print(f"Resultado: {num1} - {num2} = {restar(num1, num2)}")
            elif opcion == '3':
                print(f"Resultado: {num1} * {num2} = {multiplicar(num1, num2)}")
            elif opcion == '4':
                print(f"Resultado: {num1} / {num2} = {dividir(num1, num2)}")

            # 5. Volver al menú principal
            input("\nPresiona Enter para continuar...")
            continue

        except ValueError:
            print("⚠️ Error: Por favor ingresa solo números.")
        except ZeroDivisionError as e:
            print(f"⚠️ Error: {e}")
    else:
        print("Opción no válida, intenta de nuevo.")
```

Actividad 3: Lista de tareas Pasos:   
1. Crear una lista vacía para almacenar las tareas.
2. Definir funciones para agregar, eliminar, y mostrar tareas.
3. Implementar un menú principal usando un bucle while.
4. Usar if-elif-else para manejar las opciones del menú.
5. Utilizar un bucle for para mostrar todas las tareas.
6. Implementar try-except para manejar errores al eliminar tareas. 7. Usar break para salir del programa.

```python
# 1. Crear una lista vacía para almacenar las tareas
lista_tareas = []

# 2. Definición de funciones
def agregar_tarea(tarea):
    lista_tareas.append(tarea)
    print(f"✅ Tarea '{tarea}' agregada con éxito.")

def mostrar_tareas():
    print("\n--- TAREAS PENDIENTES ---")
    if not lista_tareas:
        print("La lista está vacía.")
    else:
        # 5. Usar un bucle for para mostrar todas las tareas
        for i, tarea in enumerate(lista_tareas, 1):
            print(f"{i}. {tarea}")

def eliminar_tarea(indice):
    # 6. Implementar try-except para manejar errores al eliminar
    try:
        tarea_eliminada = lista_tareas.pop(indice - 1)
        print(f"🗑️ Tarea '{tarea_eliminada}' eliminada.")
    except IndexError:
        print("⚠️ Error: El número de tarea no existe.")
    except ValueError:
        print("⚠️ Error: Debes ingresar el número de la posición de la tarea.")

# 3. Menú principal usando un bucle while
while True:
    print("\n--- GESTOR DE TAREAS ---")
    print("1. Agregar tarea")
    print("2. Mostrar tareas")
    print("3. Eliminar tarea")
    print("4. Salir")

    opcion = input("\nSelecciona una opción (1-4): ")

    # 4. Usar if-elif-else para manejar las opciones
    if opcion == '1':
        nueva_tarea = input("Escribe la tarea: ")
        agregar_tarea(nueva_tarea)

    elif opcion == '2':
        mostrar_tareas()

    elif opcion == '3':
        mostrar_tareas()
        if lista_tareas:
            try:
                posicion = int(input("Ingresa el número de la tarea a eliminar: "))
                eliminar_tarea(posicion)
            except ValueError:
                print("⚠️ Error: Por favor, ingresa un número válido.")

    elif opcion == '4':
        # 7. Usar break para salir del programa
        print("¡Saliendo del gestor de tareas! Que tengas un buen día, Luis Alfonso.")
        break

    else:
        print("Opción no válida. Intenta de nuevo.")
```
